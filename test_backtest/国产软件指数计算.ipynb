# 指数计算

1. 定义基期
2. 4种计算方式:
    - 基于收盘价 等权
    - 基于成交量加权
    - 基于流动市值加权
    - 基于总市值加权

```python
block=QA.QA_fetch_stock_block_adv().get_block('国产软件')
```

```python
data=QA.QA_fetch_stock_day_adv(block.code,'2018-01-01','2018-08-21')
```

```python
data
```

```python
block_mv=data.add_func(QA.QA_data_marketvalue)
```

```python
block_mv
```

```python
import numpy as np
```

```python
res=block_mv.groupby(level=0).apply(lambda x: np.average(x.close,weights=x.shares))
```

```python
resa=block_mv.groupby(level=0).apply(lambda x: np.average(x.close,weights=x.lshares))
resb=block_mv.groupby(level=0).apply(lambda x: np.average(x.close))
resc=block_mv.groupby(level=0).apply(lambda x: np.average(x.close,weights=x.volume))
```

```python
#res/res.iloc[0]*1000
```

```python
res1=res/res.iloc[0]*1000
res2=resa/resa.iloc[0]*1000
res3=resb/resb.iloc[0]*1000
res4=resc/resc.iloc[0]*1000
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 12))
plt.style.use('ggplot')
plt.title('CN Software Index')
res1.plot()
res2.plot()
res3.plot()
res4.plot()
plt.legend(['market_value','liquidity_value','close','volume'])
```
